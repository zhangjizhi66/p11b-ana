# 硅条分辨检查

```c++
const int Nrun = 8;
const int run[8] = {5402,5405,5406,5407,5408,5413,5415,5416};
```

```c++
//%jsroot on
TFile *ipf[Nrun];
for (int irun = 0; irun < Nrun; irun++)
    ipf[irun] = new TFile(Form("ENN202309_%d/6Draw/ana_%d_W1000.root",run[irun],run[irun]));
TH2I *h2;
TH1I *h1;
TCanvas *c1 = new TCanvas;
gStyle->SetOptStat(0);
```

## run5402-5408 dssd1-2 y0-7 下半部分横条分辨差            run5413-5416 变好

```c++
//%jsroot on
if (h1) delete h1;
h1 = (TH1I*)ipf[0]->Get("hs");
for (int irun = 1; irun < Nrun; irun++)
    h1->Add( (TH1I*)ipf[irun]->Get("hs") );
h1->SetTitle("Run 5402-5416");
h1->GetXaxis()->SetRangeUser(7000,18000);
h1->GetYaxis()->SetRangeUser(0,3000);
h1->GetXaxis()->SetTitle("Invariant Mass / keV");
h1->GetYaxis()->SetTitle("Count");
h1->Draw();
c1->Draw();
```

```c++
if (h2) delete h2;
h2 = (TH2I*)ipf[0]->Get("h1_16");
for (int irun = 1; irun < Nrun; irun++)
    h2->Add( (TH2I*)ipf[irun]->Get("h1_16") );
h2->SetTitle("Run 5402-5416");
h2->Draw("colz");
c1->SetLogz();
c1->Draw();
```

```c++
if (h2) delete h2;
h2 = (TH2I*)ipf[0]->Get("hsp");
for (int irun = 1; irun < Nrun; irun++)
    h2->Add( (TH2I*)ipf[irun]->Get("hsp") );
h2->SetTitle("Run 5402-5416");
h2->GetXaxis()->SetRangeUser(7000,18000);
h2->GetYaxis()->SetRangeUser(0,350);
h2->GetYaxis()->SetTitle("CM Ptot / MeV/c");
h2->GetXaxis()->SetTitle("Invariant Mass / keV");
h2->Draw("colz");
c1->SetLogz();
c1->Draw();
```
